```python
from pathlib import Path
import zipfile


def clean_text(series):
    return series.fillna("").astype(str).str.strip()


output_dir = Path("Data/schema_tables")
output_dir.mkdir(parents=True, exist_ok=True)

raw_df = pd.read_csv("Data/jobs_raw.csv")

organisations = pd.DataFrame({
    "organisation_name": clean_text(raw_df["employer"])
})
organisations = organisations[organisations["organisation_name"] != ""].drop_duplicates().reset_index(drop=True)
organisations.insert(0, "organisation_id", range(1, len(organisations) + 1))
organisations["organisation_type"] = ""
organisations["website"] = ""
organisations = organisations[["organisation_id", "organisation_name", "organisation_type", "website"]]
org_lookup = dict(zip(organisations["organisation_name"], organisations["organisation_id"]))


department_source = pd.DataFrame({
    "organisation_name": clean_text(raw_df["employer"]),
    "department_name": clean_text(raw_df["department"]),
})
department_source = department_source[
    (department_source["organisation_name"] != "") & (department_source["department_name"] != "")
].drop_duplicates().reset_index(drop=True)
department_source["organisation_id"] = department_source["organisation_name"].map(org_lookup)
department_source.insert(0, "department_id", range(1, len(department_source) + 1))
department_source = department_source[["department_id", "organisation_id", "department_name"]]


location_source = pd.DataFrame({
    "site_name": clean_text(raw_df["location"]),
    "city": clean_text(raw_df["json_address_locality"]),
    "county": "",
    "postcode": clean_text(raw_df["json_address_postcode"]).replace("", pd.NA).fillna(clean_text(raw_df["employer_postcode"])),
    "region": "",
    "country": "",
})
location_source = location_source[location_source["site_name"] != ""].drop_duplicates().reset_index(drop=True)
location_source.insert(0, "location_id", range(1, len(location_source) + 1))
location_source = location_source[["location_id", "site_name", "city", "county", "postcode", "region", "country"]]


occupation_source = pd.DataFrame({
    "category": clean_text(raw_df["staff_group"]),
    "specialty": clean_text(raw_df["specialty"]),
    "nhs_band": clean_text(raw_df["pay_band"]),
})
occupation_source = occupation_source[
    (occupation_source["category"] != "")
    | (occupation_source["specialty"] != "")
    | (occupation_source["nhs_band"] != "")
].drop_duplicates().reset_index(drop=True)
occupation_source.insert(0, "occupation_id", range(1, len(occupation_source) + 1))
occupation_source = occupation_source[["occupation_id", "category", "specialty", "nhs_band"]]


salary_source = pd.DataFrame({
    "minimum_salary": pd.to_numeric(raw_df["json_salary_min"], errors="coerce"),
    "maximum_salary": pd.to_numeric(raw_df["json_salary_max"], errors="coerce"),
    "salary_type": clean_text(raw_df["json_salary_unit"]),
    "pay_band": clean_text(raw_df["pay_band"]),
})
salary_source = salary_source.drop_duplicates().reset_index(drop=True)
salary_source.insert(0, "salary_id", range(1, len(salary_source) + 1))
salary_source = salary_source[["salary_id", "minimum_salary", "maximum_salary", "salary_type", "pay_band"]]


recruiter_source = pd.DataFrame({
    "name": clean_text(raw_df["json_hire_name"]),
})
recruiter_source = recruiter_source[recruiter_source["name"] != ""].drop_duplicates().reset_index(drop=True)
recruiter_source.insert(0, "recruiter_id", range(1, len(recruiter_source) + 1))
recruiter_source["email"] = ""
recruiter_source["phone"] = ""
recruiter_source = recruiter_source[["recruiter_id", "name", "email", "phone"]]
recruiter_lookup = dict(zip(recruiter_source["name"], recruiter_source["recruiter_id"]))


salary_lookup = {
    (row.minimum_salary, row.maximum_salary, row.salary_type, row.pay_band): row.salary_id
    for row in salary_source.itertuples(index=False)
}
occupation_lookup = {
    (row.category, row.specialty, row.nhs_band): row.occupation_id
    for row in occupation_source.itertuples(index=False)
}
location_lookup = {
    (row.site_name, row.city, row.postcode): row.location_id
    for row in location_source.itertuples(index=False)
}
department_lookup = {
    (row.organisation_id, row.department_name): row.department_id
    for row in department_source.itertuples(index=False)
}


job_source = pd.DataFrame({
    "reference_number": clean_text(raw_df["job_reference"]),
    "job_title": clean_text(raw_df["job_title"]),
    "job_description": clean_text(raw_df["full_description"]),
    "contract_type": clean_text(raw_df["job_type"]),
    "working_pattern": clean_text(raw_df["working_pattern"]),
    "employment_type": clean_text(raw_df["json_employment_type"]),
    "organisation_name": clean_text(raw_df["employer"]),
    "department_name": clean_text(raw_df["department"]),
    "site_name": clean_text(raw_df["location"]),
    "city": clean_text(raw_df["json_address_locality"]),
    "postcode": clean_text(raw_df["json_address_postcode"]).replace("", pd.NA).fillna(clean_text(raw_df["employer_postcode"])),
    "category": clean_text(raw_df["staff_group"]),
    "specialty": clean_text(raw_df["specialty"]),
    "nhs_band": clean_text(raw_df["pay_band"]),
    "salary_min": pd.to_numeric(raw_df["json_salary_min"], errors="coerce"),
    "salary_max": pd.to_numeric(raw_df["json_salary_max"], errors="coerce"),
    "salary_type": clean_text(raw_df["json_salary_unit"]),
    "recruiter_name": clean_text(raw_df["json_hire_name"]),
})
job_source = job_source[job_source["reference_number"] != ""].drop_duplicates(subset=["reference_number"]).reset_index(drop=True)
job_source.insert(0, "job_id", range(1, len(job_source) + 1))
job_source["organisation_id"] = job_source["organisation_name"].map(org_lookup)
job_source["department_id"] = [
    department_lookup.get((org_id, dept_name))
    for org_id, dept_name in zip(job_source["organisation_id"], job_source["department_name"])
]
job_source["location_id"] = [
    location_lookup.get((site, city, postcode))
    for site, city, postcode in zip(job_source["site_name"], job_source["city"], job_source["postcode"])
]
job_source["occupation_id"] = [
    occupation_lookup.get((category, specialty, band))
    for category, specialty, band in zip(job_source["category"], job_source["specialty"], job_source["nhs_band"])
]
job_source["salary_id"] = [
    salary_lookup.get((salary_min, salary_max, salary_type, pay_band))
    for salary_min, salary_max, salary_type, pay_band in zip(
        job_source["salary_min"], job_source["salary_max"], job_source["salary_type"], job_source["nhs_band"]
    )
]

job_table = job_source[[
    "job_id",
    "reference_number",
    "job_title",
    "job_description",
    "contract_type",
    "working_pattern",
    "employment_type",
    "organisation_id",
    "department_id",
    "location_id",
    "occupation_id",
    "salary_id",
]].copy()
job_table["status"] = ""
job_table = job_table[[
    "job_id",
    "reference_number",
    "job_title",
    "job_description",
    "contract_type",
    "working_pattern",
    "employment_type",
    "status",
    "organisation_id",
    "department_id",
    "location_id",
    "occupation_id",
    "salary_id",
]]

job_contact = pd.DataFrame({
    "job_id": job_source["job_id"],
    "recruiter_id": job_source["recruiter_name"].map(recruiter_lookup),
    "contact_type": "hiring_contact",
})

applicant = pd.DataFrame(columns=["applicant_id", "first_name", "last_name", "email"])
application = pd.DataFrame(columns=["application_id", "applicant_id", "job_id", "application_date", "status", "score"])

tables = {
    "organisation": organisations,
    "department": department_source,
    "location": location_source,
    "occupation": occupation_source,
    "salary": salary_source,
    "recruiter": recruiter_source,
    "job": job_table,
    "job_contact": job_contact,
    "applicant": applicant,
    "application": application,
}

written_files = []
for table_name, table_df in tables.items():
    csv_path = output_dir / f"{table_name}.csv"
    table_df.to_csv(csv_path, index=False)
    written_files.append(csv_path)

zip_path = Path("Data/schema_tables.zip")
with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as archive:
    for csv_path in written_files:
        archive.write(csv_path, arcname=csv_path.name)

print(f"Wrote {len(written_files)} table CSVs to {output_dir}")
print(f"Created archive: {zip_path}")
```
